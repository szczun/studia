```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = pd.read_csv("cw1c.csv", comment = "#")
collector_current = np.array_split(data['Ic (A)'], 6)
collector_emiter_voltage = np.array_split(data['Vce (V)'], 6)
Vrb = np.array_split(data['Vrb (V)'], 6)
thermal_voltage = 26/1000
early_voltage = 102.97553395
```

```python
r_ce_def = np.diff(collector_emiter_voltage)/np.diff(collector_current)
r_ce_lowsignal = early_voltage/collector_current[5][0:200]
r_ce_lowsignal
```

```python
for index, item in enumerate(r_ce_def):
    r_ce_lowsignal = early_voltage/collector_current[index][0:200]
    mask = np.isfinite(item) & (item >= 0)
    col_curr = collector_current[index][0:200]
    col_curr_filtered = col_curr[mask]
    r_ce_filtered = item[mask]
    
    plt.figure(index, (8, 6))
    plt.semilogy(col_curr_filtered, r_ce_filtered, marker = 'o', linestyle = ' ', color = 'black', markersize = 4, label = r"$r_{ce}$ z definicji")
    plt.semilogy(col_curr, r_ce_lowsignal, marker = 'o', linestyle = ' ', color = 'red', markersize = 4, label = "Wzór małosygnałowy")
    plt.ylabel(r"$r_{ce}$ [V] w skali logarytmicznej", fontsize = 12)
    plt.xlabel(r"$I_{C}$ [A]", fontsize = 12)
    plt.legend(fontsize = 12)
    plt.grid()
```
